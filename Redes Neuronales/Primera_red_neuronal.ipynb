```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/celsius.csv", sep = ";")
data
```

```python
caracteristica = data["Celsius"].values
etiqueta = data["Fahrenheit"].values
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape = (1,)), # Aqui estamos especificando los datos de entrada en este caso es celsius
    tf.keras.layers.Dense(units =1) # Aqui estamos especificando los datos de salida en este caso los fahrenheit
])

model.compile(
    optimizer = tf.keras.optimizers.Adam(.1), #El optimizer sirve para que la red neuronal sepa que tiene que hacer, en este caso se le pasa la tasa de aprendrizaje por repeticion
    loss = "mean_squared_error" #Esta es la perdida que va a medir
)
```

Para entrenar el modelo es similar a como se hace

```python
training = model.fit(caracteristica, etiqueta, epochs= 1000) #Esto de epochs se refiere a cuantas vueltas a dar al set de datos y junto con el optimizer ver hasta cuando aprende
```

```python
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de la epoca")
plt.plot(training.history['loss'])
```

```python
celsius = int(input("Ingrese un valor en grados celsius: "))
predict = model.predict(np.array([celsius]))
print(f"Prediccion de {celsius} celsius son {predict[0]}")
```
